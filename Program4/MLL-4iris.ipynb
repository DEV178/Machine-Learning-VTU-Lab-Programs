```python
import numpy as np
import pandas as pd
import time
iris=pd.read_csv('iris.csv')
```

```python
#X = np.array(([5.1,3.5,1.4,0.2], [4.9,3,1.4,0.2], [7,3.2,4.7,1.4],[6.4,3.2,4.5,1.5],[6.3,3.3,6,2.5],[5.8,2.7,5.1,1.9]), dtype=float)
#y = np.array(([1],[1], [2],[2],[3], [3]), dtype=float)
#y = y/100
X=iris[["sepal length","sepal width","petal length","petal width"]]
y=np.array((iris[["species"]].head(150)),dtype=float)
y=y/10
```

```python
def sigmoid (x):
    return 1/(1 + np.exp(-x))
```

```python
epoch=50000 #Setting training iterations
lr=0.1 #Setting learning rate
inputlayer_neurons = 4 #number of features in data set
hiddenlayer_neurons = 3 #number of hidden layers neurons
output_neurons = 1 #number of neurons at output layer
```

```python
wh=np.random.uniform(size=(inputlayer_neurons,hiddenlayer_neurons))
bh=np.random.uniform(size=(1,hiddenlayer_neurons))
wout=np.random.uniform(size=(hiddenlayer_neurons,output_neurons))
bout=np.random.uniform(size=(1,output_neurons))
```

```python
start_time=time.process_time()
for i in range(epoch):
#Forward Propogation
    neth1=np.dot(X,wh)+bh
    outh1 = sigmoid(neth1)
    neto1=np.dot(outh1,wout)+bout
    outo1 = sigmoid(neto1)

#Backpropagation
    deltak=(y-outo1)*(outo1)*(1-outo1)
    
    deltaj = deltak.dot(wout.T)*(outh1)*(1-outh1)
    
    wout += outh1.T.dot(deltak) *lr
    #bout += np.sum(deltak, axis=0,keepdims=True) *lr
    wh += X.T.dot(deltaj) *lr
    #bh += np.sum(deltaj, axis=0,keepdims=True) *lr
```

```python
end_time=time.process_time()-start_time
print(end_time,"seconds")
print("Input: \n" + str(X)) 
print("Actual Output: \n" + str(y))
print("Predicted Output: \n" ,outo1)
```
